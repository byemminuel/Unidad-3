## 2.1 Explorar Plazo del préstamo

### Explorar las características del plazo del préstamo
1. Mostrar los valores únicos en la columna plazo
2. Contar los valores de los diferentes plazos del crédito
3. Realizar una gráfica de barras que compare los plazos del crédito

```python
lending_club['term'].unique()
```

```python
lending_club['term'].value_counts()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, grafico = plt.subplots(figsize=(10, 5))
plt.bar(lending_club['term'].unique(), lending_club['term'].value_counts())
plt.show()
```

## 2.2 Explorar la Taza de interés

### Explorar la característica de la taza de interés
4. Calcular la tasa promedio por plazo
5. Graficar la tasa promedio
6. Relizar un histograma sobre la columna ***int_rate*** para verificar la distribución de la tasa de interés

```python
lending_club['int_rate'].groupby(lending_club['term']).mean()
```

```python
fig, grafico2 = plt.subplots(figsize=(10, 5))
plt.bar(lending_club['term'].unique(), lending_club['int_rate'].groupby(lending_club['term']).mean())
plt.show()
```

```python
fig, grafico3 = plt.subplots(2, 2, figsize=(20, 10))
grafico3[0][0].hist(lending_club['int_rate'])
grafico3[0][0].set_xlabel('Tasa de interés', fontsize=16)
grafico3[0][0].tick_params(labelsize=14)
grafico3[0][1].set_visible(False)
grafico3[1][0].set_visible(False)
grafico3[1][1].set_visible(False)
```

# 2.3 Explorar la calificación del préstamo

7. ¿Cuántas calificaciones de préstamo diferentes hay en el conjunto de datos?
8. Graficar con barras los valores obtenidos

```python
len(lending_club['sub_grade'].unique())
```

```python
fig, grafico4 = plt.subplots(figsize=(10, 5))
plt.bar(lending_club['sub_grade'].unique(), lending_club['int_rate'].groupby(lending_club['sub_grade']).mean())
grafico4.set_xlabel('Calificación de préstamo', fontsize=16)
grafico4.set_ylabel('Tasa de interés promedio', fontsize=16)
plt.show()
```

# 2.4 Explorar los Préstamos por Estado

9. Contabilizar el número de préstamos por estado
10. Graficar con barras horizontales los valores obtenidos

```python
lending_club['addr_state'].value_counts()
```

```python
fig, grafico4 = plt.subplots(figsize=(10, 5))
plt.barh(lending_club['addr_state'].unique(), lending_club['addr_state'].value_counts())
grafico4.set_xlabel('Estado', fontsize=16)
grafico4.set_ylabel('Cantidad de préstamos', fontsize=16)
plt.show()
```
